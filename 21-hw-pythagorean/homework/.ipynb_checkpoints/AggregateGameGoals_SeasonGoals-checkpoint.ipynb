```python
import pandas
```

```python
dm = pandas.read_csv('2010game_physical.csv')
```

```python
dm['awayWin'] = dm.apply(lambda x: 1 if x['awayTeam']==x['winteamcode'] else 0, 1)
dm['homeWin'] = dm.apply(lambda x: 1 if x['awayTeam']!=x['winteamcode'] else 0, 1)

dm.head()
```

```python
da = dm.groupby(['awayTeam'], as_index=False)['awayGoals', 'homeGoals', 'awayWin', 'homeWin'].agg('sum')
da = da.rename(columns={'awayTeam': 'Team', 'awayGoals': 'GoalsFor', 'homeGoals': 'GoalsAllowed', 'awayWin': 'Wins', 'homeWin': 'Losses'})

dh = dm.groupby(['homeTeam'], as_index=False)['awayGoals', 'homeGoals', 'awayWin', 'homeWin'].agg('sum')
dh = dh.rename(columns={'homeTeam': 'Team', 'awayGoals': 'GoalsAllowed', 'homeGoals': 'GoalsFor', 'homeWin': 'Wins', 'awayWin': 'Losses'})

ds = da.append(dh)
ds = ds[['Team', 'GoalsFor', 'GoalsAllowed', 'Wins', 'Losses']]
ds = ds.groupby(['Team'], as_index=False).sum()
ds['GamesPlayed'] = ds.Wins + ds.Losses
ds = ds[['Team', 'GamesPlayed', 'GoalsFor', 'GoalsAllowed', 'Wins', 'Losses']]
ds.to_csv('NHL_season_team_goals.csv')
```
